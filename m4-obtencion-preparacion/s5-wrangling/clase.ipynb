```python
import pandas as pd

df=pd.read_csv("base_ventas_inconsistente.csv")
```

```python
df.sample(10)
```

## Remplazo

```python
print(df.loc[107])
df['Pais']=df['Pais'].replace('MICRONESIA','Micronesia')
print()
print(df.loc[107])
```

## Lambda

```python
df['Pais']=df['Pais'].apply(lambda x:x.capitalize())
df.sample(15)
```

## Agrupamiento

```python
df['Grupo']=pd.cut(df['Precio_Total'],bins=[0,1000,3000,5000],labels=['Pequeña','Mediana','Grande'])
df.sample(15)
```

## Pivot (redimensionar)

```python
df2=pd.DataFrame({
    'Fecha':['2025-01','2025-01','2025-02','2025-02'],
    'Producto':['A','B','A','B'],
    'Ventas':[100,150,120,160]
})
print(df2)
df2=df2.pivot(index='Fecha',columns='Producto',values='Ventas')
print(f"\n{df2}")
```
