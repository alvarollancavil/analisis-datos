```python
def resumen(df):
    columnas_numericas = df.select_dtypes(include=['number']).columns
    nombres = []
    minimo = []
    maximo = []
    media = []

    for x in columnas_numericas:
        nombres.append(x)
        minimo.append(df[x].min())
        maximo.append(df[x].max())
        media.append(df[x].mean())

    df_resumen = pd.DataFrame({
        'Variable': nombres,
        'Mínimo': minimo,
        'Máximo': maximo,
        'Media': media
    })
    
    return df_resumen
```

```python
def obtener_emisiones(data, estacion, magnitud, fecha_inicio, fecha_fin, 
                        col_estacion='ESTACION', col_fecha='FECHA',col_magnitud='MAGNITUD', col_emisiones='CONTAMINANTES'):
  
    estaciones_validas = data[col_estacion].unique()
    magnitudes_validas = data[col_magnitud].unique()

    if estacion not in estaciones_validas:
        raise ValueError(f"Estación '{estacion}' no encontrada. Estaciones válidas: {', '.join(map(str, estaciones_validas))}")
    
    if magnitud not in magnitudes_validas:
        raise ValueError(f"Magnitud '{magnitud}' no encontrada. Magnitudes válidas: {', '.join(map(str, magnitudes_validas))}")
    
    try:
        fecha_inicio_dt = pd.to_datetime(fecha_inicio)
        fecha_fin_dt = pd.to_datetime(fecha_fin)
    except ValueError as e:
        raise ValueError(f"Formato de fecha inválido. Use 'YYYY-MM-DD'. Error: {str(e)}")
    
    if fecha_inicio_dt > fecha_fin_dt:
        raise ValueError(f"Fecha de inicio {fecha_inicio} debe ser anterior o igual a fecha fin {fecha_fin}")
    
    data[col_fecha] = pd.to_datetime(data[col_fecha])
    min_fecha = data[col_fecha].min()
    max_fecha = data[col_fecha].max()
    
    if fecha_inicio_dt > max_fecha or fecha_inicio_dt < min_fecha or fecha_fin_dt > max_fecha or fecha_fin_dt < min_fecha:
        raise ValueError(f"Rango de fechas fuera de los datos disponibles. Rango en datos: {min_fecha.date()} a {max_fecha.date()}")
    
    filtro = df[(data[col_estacion]==estacion)&(data[col_magnitud]==magnitud)&(data[col_fecha]>=fecha_inicio_dt)&
                (data['FECHA']<=fecha_fin_dt)]
    filtro.reset_index(drop=True, inplace=True)
    filtro=filtro[[col_estacion,col_magnitud,col_fecha,col_emisiones]] 
    filtro=filtro.rename(columns={col_emisiones:'EMISIONES'})
    return filtro
```

```python
def resumen_emisiones_agrupadas(datos, col_agrupada='CATEGORIAS', col_emision='EMISION'):
    import pandas as pd
    import numpy as np
    from scipy.stats import kurtosis

    
    columnas_requeridas = {col_agrupada, col_emision}
    if not columnas_requeridas.issubset(datos.columns):
        faltantes = columnas_requeridas - set(datos.columns)
        raise ValueError(f"Columnas faltantes en los datos: {', '.join(faltantes)}")
    
    if not pd.api.types.is_numeric_dtype(datos[col_emision]):
        raise ValueError(f"La columna '{col_emision}' debe contener valores numéricos")
    
    def cv(x):
        return np.std(x, ddof=1) / np.mean(x) * 100 if np.mean(x) != 0 else np.nan
    
    def kurt(x):
        return kurtosis(x, fisher=True, bias=False)
    
    resumen = datos.groupby(col_agrupada)[col_emision].agg(
        N='count',
        Mínimo='min',
        Máximo='max',
        Media='mean',
        Mediana='median',
        Desv_Estándar='std',
        Varianza='var',
        Sesgo='skew',
        Curtosis=kurt,
        Percentil_25=lambda x: x.quantile(0.25),
        Percentil_75=lambda x: x.quantile(0.75),
        IQR=lambda x: x.quantile(0.75) - x.quantile(0.25),
        Coef_Variación=cv 
    )
    
    # Reordenar columnas para mejor presentación
    column_order = [
        'N', 'Mínimo', 'Percentil_25', 'Mediana', 'Media', 
        'Percentil_75', 'Máximo', 'IQR', 'Desv_Estándar', 
        'Varianza', 'Coef_Variación', 'Sesgo','Curtosis'
    ]
    
    return resumen[column_order].round(2)
```

```python
resumen_emisiones_agrupadas(df, 'MAGNITUD', 'CONTAMINANTES')
```

```python
def emisiones_agrupadas_filtro(data, estacion, magnitud, col_estacion='ESTACION', col_magnitud='MAGNITUD', col_emisiones='CONTAMINANTES'):
    estaciones_validas = data[col_estacion].unique()
    magnitudes_validas = data[col_magnitud].unique()

    if estacion not in estaciones_validas:
        raise ValueError(f"Estación '{estacion}' no encontrada. Estaciones válidas: {', '.join(map(str, estaciones_validas))}")
    
    if magnitud not in magnitudes_validas:
        raise ValueError(f"Magnitud '{magnitud}' no encontrada. Magnitudes válidas: {', '.join(map(str, magnitudes_validas))}")
    
    filtro = df[(data[col_estacion]==estacion)&(data[col_magnitud]==magnitud)]
    filtro.reset_index(drop=True, inplace=True)
    filtro=filtro[[col_estacion,col_magnitud,col_emisiones]] 
    filtro=filtro.rename(columns={col_emisiones:'EMISIONES'})

    return filtro.groupby(col_estacion).apply(resumen_emisiones_agrupadas, col_agrupada=col_magnitud, col_emision='EMISIONES')
    
    
    
```

```python
def emision_media_mensual_agrupada(data, estacion, magnitud, año, col_estacion='ESTACION',
                                   col_magnitud='MAGNITUD', col_fecha='FECHA',
                                   col_emisiones='CONTAMINANTES'):
    estaciones_validas = data[col_estacion].unique()
    magnitudes_validas = data[col_magnitud].unique()
    años_validos = data[col_fecha].dt.year.unique()

    if estacion not in estaciones_validas:
        raise ValueError(f"Estación '{estacion}' no encontrada. Estaciones válidas: {', '.join(map(str, estaciones_validas))}")
    
    if magnitud not in magnitudes_validas:
        raise ValueError(f"Magnitud '{magnitud}' no encontrada. Magnitudes válidas: {', '.join(map(str, magnitudes_validas))}")
    
    if año not in años_validos:
        raise ValueError(f"Año '{año}' no encontrado. Años válidos: {', '.join(map(str, años_validos))}")
 
    data[col_fecha] = pd.to_datetime(data[col_fecha])
    filtro = df[(data[col_estacion]==estacion)&(data[col_magnitud]==magnitud) & (df[col_fecha].dt.year == año)]
    filtro.reset_index(drop=True, inplace=True)
    filtro=filtro[[col_fecha, col_estacion, col_magnitud, col_emisiones]] 
    filtro=filtro.rename(columns={col_emisiones:'EMISIONES'})
    filtro = filtro.set_index(col_fecha)
    filtro = filtro['EMISIONES'].resample('ME').mean()
    filtro = filtro.reset_index()
    meses_espanol = {'January': "Enero", 'February': "Febrero", 'March': "Marzo", 'April': "Abril",
                     'May': "Mayo", 'June': "Junio", 'July': "Julio", 'August': "Agosto",
                     'September': "Septiembre", 'October': "Octubre", 'November': "Noviembre",
                     'December': "Diciembre"}
    filtro[col_fecha] = filtro[col_fecha].apply(dt.datetime.strftime, format='%B')
    filtro[col_fecha] = filtro[col_fecha].map(meses_espanol)
    filtro = filtro.rename(columns={col_fecha:'MES'})
    return print(f'Media mensual de emisiones para la estación "{estacion}" y magnitud "{magnitud}" del año {año}:\n{filtro}')

    

    
    
```

```python
emision_media_mensual_agrupada(df, 4, 7, 2017, col_estacion='ESTACION', col_fecha='FECHA',
                               col_emisiones='CONTAMINANTES', col_magnitud='MAGNITUD')
```

```python
def emision_media_mensual_estacion(data, estacion, col_estacion='ESTACION',
                                   col_magnitud='MAGNITUD', col_fecha='FECHA',
                                   col_emisiones='CONTAMINANTES'):
    estaciones_validas = data[col_estacion].unique()

    if estacion not in estaciones_validas:
        raise ValueError(f"Estación '{estacion}' no encontrada. Estaciones válidas: {', '.join(map(str, estaciones_validas))}")
    
    data[col_fecha] = pd.to_datetime(data[col_fecha])
    filtro = df[(data[col_estacion]==estacion)]
    filtro.reset_index(drop=True, inplace=True)
    filtro=filtro[[col_fecha, col_estacion, col_magnitud, col_emisiones]] 
    filtro=filtro.rename(columns={col_emisiones:'EMISIONES'})
    filtro = filtro.set_index(col_fecha)
    filtro = filtro.groupby(col_magnitud)['EMISIONES'].resample('ME').mean()
    filtro = filtro.reset_index()
    meses_espanol = {'January': "Enero", 'February': "Febrero", 'March': "Marzo", 'April': "Abril",
                     'May': "Mayo", 'June': "Junio", 'July': "Julio", 'August': "Agosto",
                     'September': "Septiembre", 'October': "Octubre", 'November': "Noviembre",
                     'December': "Diciembre"}
    filtro['MES'] = filtro[col_fecha].apply(dt.datetime.strftime, format='%B')
    filtro['MES'] = filtro['MES'].map(meses_espanol)
    filtro['AÑO'] = filtro[col_fecha].apply(dt.datetime.strftime, format='%Y')
    return print(f'Media mensual de emisiones para la estación "{estacion}":\n{filtro[['AÑO', 'MES', col_magnitud, 'EMISIONES']]}')
```
